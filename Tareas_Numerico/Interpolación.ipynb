```python
import numpy as np
import scipy.interpolate as inter
import matplotlib.pyplot as plt
np.set_printoptions(formatter={'float': '{: 0.4f}'.format})
```

Tomamos la funcion $f(x)=x(x-2\pi)e^{-x}$ en una muestra de varios puntos e aplicuemos las diferentes interpolaciones.

```python
x=np.arange(0,9,1)
y=x*(x-2*np.pi)*np.exp(-x)
```

Tenemos entonces la siguiente tabla de datos.

```python
for i in range(len(x)):
  print('| {:4} \t| {:2.4f}|'.format(x[i],y[i]) )
```

```python

t=np.arange(0,8,.01);

fl=inter.interp1d(x,y, kind='linear')
vl=fl(t)

fn=inter.interp1d(x,y, kind='nearest')
vn=fn(t)

fz=inter.interp1d(x,y, kind='zero')
vz=fz(t)

fc=inter.interp1d(x,y, kind='cubic')
vc=fc(t)

plt.figure(figsize=(15,10))
plt.scatter(x,y,s=140)
plt.plot(t,t*(t-2*np.pi)*np.exp(-t), '-.m',linewidth=5 )
plt.plot(t,vl,linewidth=2,color='red')
plt.plot(t,vn,linewidth=2,color='green')
plt.plot(t,vz,linewidth=2,color='cyan')
plt.plot(t,vc,linewidth=2,color='black')
plt.show()
```

```python

s=np.arange(1,14,0.1)

icx=inter.interp1d(t,coorx, kind='linear')
curvex=icx(s)

icy=inter.interp1d(t,coory, kind='cubic')
curvey=icy(s)


plt.figure(figsize=(8,8))
plt.plot(curvex,curvey,linewidth=10,color='blue',alpha=0.8)
plt.scatter(coorx,coory,s=400,color='red')
plt.axis('equal')
plt.grid(color='gray', linestyle='--',linewidth=2)
plt.show()
```
